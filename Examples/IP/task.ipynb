```python
import numpy as np
```

```python
img = np.array([[170, 238, 85, 255, 221,0],
                [68, 136, 17, 170, 119,68],
                [221, 0, 238, 136, 0,255],
                [119, 255, 85, 170, 136,238],
                [238, 17, 221, 68, 119,255],
                [85, 170, 119, 221, 17,136]])

```
